## Matplotlib

[Matplotlib](http://matplotlib.org/) is the most common charting package, see its [documentation](http://matplotlib.org/api/pyplot_api.html) for details, and its [examples](http://matplotlib.org/gallery.html#statistics) for inspiration.

### Line Plots

```
import matplotlib.pyplot as plt
 
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]
plt.plot(x, y1, "r--", label="line L")
plt.plot(x, y2, label="line H")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Example")
plt.legend()
plt.show()
```

```
import numpy as np
import pylab as pl

# 
x = np.arange(0.0, 2.0*np.pi, 0.01)
y = np.sin(x)						
z = np.cos(x)						

#
pl.plot(x, y, label='sin()')
pl.plot(x, z, label='cos()')
pl.xlabel('x')		
pl.ylabel('y')
pl.title('sin-cos') 


pl.legend(loc='center')	
#loc 
#pl.legend(loc='upper right')	
#pl.legend(loc='lower right')	
#pl.legend()				
pl.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# 
x= np.linspace(0, 2*np.pi, 500)	
y1 = np.sin(x)					
y2 = np.cos(x)
y3 = np.sin(x*x)

#
plt.figure(1)					#
#create three axes
ax1 = plt.subplot(2,2,1)			#
ax2 = plt.subplot(2,2,2)			#
ax3 = plt.subplot(2,1,2)			#

##
plt.sca(ax1)					#ax1
plt.plot(x,y1,color='red')		#
plt.ylim(-1.2,1.2)				#y

##
plt.sca(ax2)					#ax2
plt.plot(x,y2,'b--')			#
plt.ylim(-1.2,1.2)

##
plt.sca(ax3)					#ax3
plt.plot(x,y3,'g--')
plt.ylim(-1.2,1.2)

##
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d


# 
x,y = np.mgrid[-2:2:20j, -2:2:20j]
z = 50 * np.sin(x+y)		#

#
ax = plt.subplot(111, projection='3d')	#
ax.plot_surface(x,y,z,rstride=2, cstride=1, cmap=plt.cm.Blues_r)
ax.set_xlabel('X')			#
ax.set_ylabel('Y')
ax.set_zlabel('Z')


plt.show()
```
